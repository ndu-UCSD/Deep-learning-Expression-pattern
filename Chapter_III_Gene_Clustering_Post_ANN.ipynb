```python
def K_binary(data_dist,tau): # threshold for connectivity tau in [0,1]
    temp = np.where(data_dist>tau,1,data_dist)
    return sum(np.where(data_dist<tau,0,temp))

def Finding_tau(data_dist,tau = 0.4, bins = 20): # test tau starting from 0.5
    tau_tested = []
    GOF = []
    mean_K = []
    for test_tau in np.linspace(tau,0.95,10):
        tau_tested.append(test_tau)
        K_unweighted = K_binary(data_dist,test_tau)
        mean_K.append(np.mean(K_unweighted))
        hist_data= plt.hist(K_unweighted,bins = bins)
        p_K = hist_data[0]/sum(hist_data[0])
        fitting = linregress(np.log10(hist_data[1][1:]),np.log10(p_K))
        GOF.append(fitting.rvalue**2)
        plt.close()
    return tau_tested,GOF,mean_K
```

```python
# Getting Mean sq data for analysis
from sklearn.decomposition import PCA,TruncatedSVD

PCA_data =  TPM
Sample_ids = ['WT_LL_1','WT_LL_2','WT_LL_3','WT_HL_1','WT_HL_2','WT_HL_3','LHCX_LL_1','LHCX_LL_2','LHCX_LL_3','LHCX_HL_1','LHCX_HL_2','LHCX_HL_3']
PCA_data.columns = Sample_ids

data_array = []
for i in PCA_data.columns:
    data_array.append(list(PCA_data[i]))
data_array = np.array(data_array)

### Conducting PCA analysis ###
# Calculated PCA
# from sklearn.preprocessing import StandardScaler
# data_array = StandardScaler().fit_transform(data_array)
pca = PCA(n_components=2)
PCA_result = pca.fit_transform(data_array)
Ratio = pca.explained_variance_ratio_

# Calculate eigenvectors
mu = PCA_result.mean(axis=0)
PCA_result = PCA_result - mu
eigenvectors, eigenvalues, V = np.linalg.svd(PCA_result.T, full_matrices=False)
projected_data = np.dot(PCA_result, eigenvectors)
Scalar = projected_data.std(axis=0).mean()

# Plot PCA results
# sns.set(context='paper', style='white', palette='dark', font='sans-serif', font_scale=1.8)#, color_codes=color_code, rc=None)
# sns.set_style("ticks", {"xtick.major.size": 8, "ytick.major.size": 8})
# color_scale = ['#F16A70','#B1D877','#8CDCDA','#4D4D4D']
color_scale = ['gray','black','gray','black']
marker_symbol = ['^','s','*','o']
x = [xx[0] for xx in PCA_result]
y = [xx[1] for xx in PCA_result]
fig,ax = plt.subplots(1,1,figsize=(5, 5),dpi = 100)
for ii in range(4):
    ax.scatter(x[0+ii*3:3+ii*3],y[0+ii*3:3+ii*3],color = color_scale[ii],s=50,marker=marker_symbol[ii])
plt.legend(['WT 0 hour','WT 24 hour','LHCX1A 0 hour','LHCX1A 24 hour'])#,bbox_to_anchor=(1.35,1))
lim1 = 20000 #max(ax.get_xlim()[1],ax.get_ylim()[1])
lim0 = -20000 #min(ax.get_xlim()[0],ax.get_ylim()[0])
ax.set_xlim([lim0,lim1])
ax.set_ylim([lim0,lim1])
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')

for axis in eigenvectors:
    start, end = mu, mu + Ratio*Scalar*2* axis
    ax.annotate(
        '', xy=end, xycoords='data',
        xytext=start, textcoords='data',
        arrowprops=dict(facecolor='gray', width=2.0))
    ax.annotate("{0:.0f}%".format(Ratio[int(abs(axis[1]))]*100),xy=end, xytext=end+800)
ax.tick_params(
    axis='both',          # changes apply to the both-axis
    which='both',      # both major and minor ticks are affected
    #bottom='off',      # ticks along the bottom edge are off
    top='off',        # ticks along the top edge are off
    #left = 'off', 
    #labelbottom='off',
    #labelleft = 'off',
    ) # labels along the bottom edge are off
```
